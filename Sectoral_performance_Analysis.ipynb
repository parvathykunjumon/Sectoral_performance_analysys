## 📘 Introduction

The goal is to extract meaningful insights from raw economic data and understand trends, sector contributions, and investment patterns.

# Data Pre-Processiong
○ Handling missing values

○ Removing duplicates 

○ Correcting data types 

○ Creating derived columns 

○ Filtering or aggregating data 


### Handling missing value

```python

df['industry'] = df['industry'].fillna("Unknown")
df['institutional_sector'] = df['institutional_sector'].fillna("Unknown")
df['subindustry'] = df['subindustry'].fillna("Not Available")
df['quarter'] = df['quarter'].fillna("Annual")

df['current_price'] = df['current_price'].fillna(df['current_price'].median())
df['constant_price'] = df['constant_price'].fillna(df['constant_price'].median())
df.to_csv("sectoral_cleaned_data.csv", index=False)
```

```python
df['revision'] = df['revision'].fillna("Not Available")
```

```python
df.isnull().sum()
```

### clean unit value in the column 'Unit'

```python

df['unit'] = df['unit'].replace({
    '₹ Crore': 'INR Crore',
    'Rs Crore': 'INR Crore'
})
df.to_csv("sectoral_cleaned_data.csv", index=False)
```

### Remove duplicates

```python

df = df.drop_duplicates() #57 rows affected
```

```python
#Fix column name
df.columns = df.columns.str.strip().str.upper().str.replace(" ", "_")
```

```python
df.head()
```

### Handle invalid value

```python

df = df[df['CURRENT_PRICE'] >= 0]
```

## Data type check

```python
df.dtypes
```

## Correcting Data type

### 1. Convert Categorical column

```python
categorical_cols = [
    'BASE_YEAR', 'SERIES', 'INDICATOR', 'FREQUENCY',
    'REVISION', 'INDUSTRY', 'SUBINDUSTRY',
    'INSTITUTIONAL_SECTOR', 'QUARTER'
]

for col in categorical_cols:
    df[col] = df[col].astype('category')
```

### 2. Handling Year (Derived column)

```python
df['YEAR_NUM'] = df['YEAR'].str[:4].astype(int)
```

```python
df['BASE_YEAR_NUM'] = df['BASE_YEAR'].str[:4].astype(int)
```

```python
# Final check
df.dtypes
```

## Derived columns
 🔹 1. Growth Rate (Shows how economy/sector is growing)

```python
df['GROWTH_RATE'] = df.groupby('INDUSTRY', observed=False)['CONSTANT_PRICE'].pct_change() * 100
```

* Handling nulll value on Growth_rate  
For each INDUSTRY, the first year has:
No previous value → so growth cannot be calculated

```python

df['GROWTH_RATE'] = df['GROWTH_RATE'].fillna(0)
```

🔹 2. Price Difference (Inflation Impact)

```python
df['PRICE_DIFF'] = df['CURRENT_PRICE'] - df['CONSTANT_PRICE']
```

🔹 3. Price Ratio 

```python
df['PRICE_RATIO'] = df['CURRENT_PRICE'] / df['CONSTANT_PRICE']
```

🔹 4. Sector Contribution (%)

```python
total_per_year = df.groupby('YEAR_NUM')['CONSTANT_PRICE'].transform('sum')

df['SECTOR_CONTRIBUTION_%'] = (df['CONSTANT_PRICE'] / total_per_year) * 100
```

```python
df.to_csv("sectoral_cleaned_data.csv", index=False)
```

# 📊 5. Insight Generation and Report

## 📝 Approach & Methodology

The dataset was first preprocessed by handling missing values, correcting data types, and creating derived features such as growth rate and sector contribution. Exploratory Data Analysis (EDA) was then performed using univariate, bivariate, and multivariate techniques. Statistical summaries and visualizations were used together to identify patterns, trends, and relationships in the data.

## 📈 Key Insights
🔹 1. Overall Economic Growth

The analysis shows a consistent upward trend in constant price over the years, indicating steady economic growth. This is supported by increasing yearly averages and visualized through line charts.

🔹 2. Sector-wise Contribution

A few industries contribute significantly more to the economy compared to others. Aggregation results and bar charts highlight dominant sectors, indicating uneven distribution of economic activity.

🔹 3. Growth Rate Variability

Growth rates vary considerably across industries. Some sectors show stable growth, while others exhibit high fluctuations, as seen in box plots and supported by higher standard deviation values.

🔹 4. Relationship Between Economic Indicators

There is a strong positive correlation between current price and constant price, indicating consistent economic scaling. However, growth rate shows weaker correlation, suggesting independent variation across sectors.

🔹 5. Production vs Investment Patterns

GVA (production) and GFCF (investment) show different trends across industries. This indicates that investment is not uniform and varies depending on sector performance.

🔹 6. Patterns & Anomalies

Certain industries show sudden spikes or drops in growth rate, indicating possible economic shocks, policy changes, or sector-specific dynamics.

### 📊 Statistical Support

Statistical summaries such as mean, standard deviation, and range were used to validate insights. Industry-wise and year-wise summaries confirmed sector dominance, variability, and overall economic growth trends observed in visualizations.

### 🔍 Interpretation of Results

The economy demonstrates a general growth trend, but the contribution and stability vary significantly across industries. While some sectors consistently drive economic output, others experience volatility, highlighting the importance of sector-specific analysis.

## 🎯 Conclusion

This analysis highlights steady economic growth over time, with strong contributions from key industries. However, variability in growth rates suggests that not all sectors perform uniformly. The relationship between current and constant prices indicates stable economic scaling, while differences in production and investment trends provide deeper insights into economic structure.
